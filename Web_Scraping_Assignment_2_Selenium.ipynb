# 1Question: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location.Scrape the job-title, job-location, company_name, experience_required.Scrape first 10 jobs data.

```python
driver=webdriver.Chrome('chromedriver.exe')
```

```python
url='https://www.naukri.com/'
driver.get(url)
```

```python
search_button=driver.find_element_by_id('qsb-keyword-sugg')
search_button.send_keys('Data Analyst')

search_loc=driver.find_element_by_id('qsb-location-sugg')
search_loc.send_keys('Bangalore')


search=driver.find_element_by_xpath("//button[@class='btn']")
search.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
exp_req=[]
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    job_title.append(i.text)
for i in driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']//span"):
    job_location.append(i.text)
for i in driver.find_elements_by_xpath("//div[@class='mt-7 companyInfo subheading lh16']//a"):
    company_name.append(i.text)
for i in driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']//span"):
    exp_req.append(i.text)
```

```python
df=pd.DataFrame({})
df['Job_title']=job_title[:10]
df['job_location']=job_location[:10]
df['company']=company_name[:10]
df['experience']=exp_req[:10]
```

# Q10: Go to webpage https://www.amazon.in/
Enter “Laptop” in the search field and then click the search icon.
Then set CPU Type filter to “Intel Core i7” and “Intel Core i9” as shown in the below image:

```python
url='https://www.amazon.in/'
driver.get(url)
```

```python
search_button=driver.find_element_by_id('twotabsearchtextbox')
search_button.send_keys('Laptop')

```

```python
apply=driver.find_element_by_xpath("//div[@class='nav-search-submit nav-sprite']")
apply.click()
```

```python
filter_button=driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span")
for i in filter_button:
    if i.text=='Intel Core i7':
        i.click()
        break
```

```python
filter_button=driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span")
for i in filter_button:
    if i.text=='Intel Core i9':
        i.click()
        break
```

```python
Title=[]
Ratings=[]
price=[]
titles=driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")
for i in titles[:10]:
    Title.append(i.text)
```

```python
prices=driver.find_elements_by_xpath("//span[@class='a-price-whole']")
for i in prices[:10]:
    price.append(i.text)
```

```python
urls=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
UR=[]
for i in urls[:10]:
    UR.append(i.get_attribute('href'))
for url in UR:
    driver.get(url)
    try:                  
        rate=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        rate.click()                                                      
        rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        Ratings.append(rating.text)
        
    except NoSuchElementException   as e:
        Ratings.append("NO rating")
```

```python
df=pd.DataFrame({'Title':Title,
                'Price':price,
                'Ratings':Ratings})
```
